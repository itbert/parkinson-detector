```python
correlation = data.drop(columns='name').corr()
correlation
```

```python
plt.figure(figsize=(15, 12))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation matrix')
plt.show
```

```python
df = data
X = df[['MDVP:Fo(Hz)', 'status']]
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X)
df['Cluster'] = kmeans.labels_
plt.scatter(df['MDVP:Fo(Hz)'], df['status'], c=df['Cluster'], cmap='viridis')
plt.xlabel('MDVP')
plt.ylabel('Status Parkinson')
plt.title('Binary K-Means')
plt.show()
```
